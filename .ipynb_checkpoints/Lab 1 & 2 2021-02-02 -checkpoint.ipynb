```python
def corr_matrix(df):
    import matplotlib.pyplot as plt
    import seaborn as sns
    
    corr_matrix=df.corr(method='pearson')  # default
    fig, ax = plt.subplots(figsize=(10, 8))
    ax = sns.heatmap(corr_matrix, annot=True)
    plt.show()
```
