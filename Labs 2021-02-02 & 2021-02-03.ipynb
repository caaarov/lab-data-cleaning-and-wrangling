```python
def model_function(df, target, type_model, test_size=0.3, multi_class="multinomial", k=5, n=2): #target has to be string
    
    #x-y-split
    Y=df[target]
    X=df.drop([target],axis=1)

    #train-test-split
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size, random_state=100)
    
    #LM
    if type_model=="linear_model":
        from sklearn import linear_model
        
        lm = linear_model.LinearRegression()
        model = lm.fit(X_train,y_train)
        predictions  = model.predict(X_test)
        predictions_df = pd.DataFrame(predictions,columns=[target])
    
    #LOG
    elif type_model=="log_model":
        from sklearn.linear_model import LogisticRegression
        from sklearn.metrics import cohen_kappa_score
        
        model = LogisticRegression(random_state=0, solver='lbfgs', multi_class=multi_class).fit(X_train, y_train)
        predictions  = model.predict(X_test)
        predictions_df = pd.DataFrame(model.predict(X_test),columns=[target])
        
        print("The accuracy of the model in the TEST set is: ", round(model.score(X_test, y_test),2))
        print("The kappa of the model is: ", cohen_kappa_score(Y,model.predict(X)))
    
    #KNN
    elif type_model=="knn_model":
        from sklearn.neighbors import KNeighborsRegressor
        
        model = KNeighborsRegressor(n_neighbors=k) # The keyword "n_neighbors" is what sets the K.
        model = model.fit(X_train, y_train)
        predictions  = model.predict(X_test)
        predictions_df = pd.DataFrame(predictions,columns=[target])
    
    #MLP
    elif type_model=="mlp_model":
        from sklearn.neural_network import MLPRegressor

        X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=100)

        model = MLPRegressor(random_state=1, max_iter=500).fit(X_train, y_train)
        predictions = model.predict(X_test) #array
        predictions_df = pd.DataFrame(predictions,columns=[target]) 
    
    #POLY
    elif type_model=="poly_model":
        from sklearn.linear_model import Ridge
        from sklearn.preprocessing import PolynomialFeatures
        from sklearn.pipeline import make_pipeline
        
        model = make_pipeline(PolynomialFeatures(n), Ridge()).fit(X_train, y_train)
        predictions = model.predict(X_test)
        predictions_df = pd.DataFrame(predictions,columns=[target]) 
    
    #get the error-matrix
    from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
    import math
    
    mae = mean_absolute_error(y_test,predictions)
    mse = mean_squared_error(y_test,predictions)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test,predictions)
    
    N = len(X_test)
    p = X_test.shape[1]
    adj_r2 = 1-((1-r2)*(N-1)/(N-p-1))

    print("The mean absolute error of the model in the test set is: %6.2f" % (mae))
    print("The mean squared error of the model in the test set is: %6.2f" % (mse))
    print("The root mean squared error of the model in the test set is: %6.2f" % (rmse))
    print("The R2 of the model in the test set is: %4.2f" % (r2))
    print("The adjusted R2 of the model in the test set is: %4.2f" % (adj_r2))
    
    results_error_matrix=[mae,mse,rmse,r2,adj_r2]
    
    sns.scatterplot(y = predictions, x = y_test)
    
    return model, predictions, predictions_df, results_error_matrix
```

```python
def train_test_model(model, y_target, x_data, random_seed):
    
    #traintest split
    X_train, X_test, y_train, y_test = train_test_split(x_data, y_target, test_size=0.3, random_state=random_seed)
    #fitmodel
    model = model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    predictions_df = pd.DataFrame(predictions,columns=["target"])
    
    #get the error-matrix
    from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
    import math
    
    mae = mean_absolute_error(y_test,predictions)
    mse = mean_squared_error(y_test,predictions)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test,predictions)
    
    N = len(X_test)
    p = X_test.shape[1]
    adj_r2 = 1-((1-r2)*(N-1)/(N-p-1))

    print("The mean absolute error of the model in the test set is: %6.2f" % (mae))
    print("The mean squared error of the model in the test set is: %6.2f" % (mse))
    print("The root mean squared error of the model in the test set is: %6.2f" % (rmse))
    print("The R2 of the model in the test set is: %4.2f" % (r2))
    print("The adjusted R2 of the model in the test set is: %4.2f" % (adj_r2))
    
    results_error_matrix=[mae,mse,rmse,r2,adj_r2]
    
    sns.scatterplot(y = predictions, x = y_test)
    
    return model, predictions, predictions_df,results_error_matrix
```
